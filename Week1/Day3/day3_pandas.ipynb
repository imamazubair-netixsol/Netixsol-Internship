# Lets move onto the 8 Business Questions
### Q1. Which 10 games generated the highest global sales?

```python
top_games = cleaned_df[["Name", "Global_Sales"]].sort_values(by="Global_Sales", ascending=False).head(10)
```

```python
top_games
```

Wii Sports is the highest selling game in the dataset, followed by Super Mario Bros and Mario Kart Wii

### Q2. If a publisher could only develop for one historical platform, which platform would have offered the largest market opportunity?

```python
platform_sales = (cleaned_df.groupby("Platform")["Global_Sales"].sum().sort_values(ascending=False))
```

```python
platform_sales
```

PS2 generated the highest lifetime sales among all gaming platforms. -w-

### Q3. Which game genre is most profitable globally? 
interesting question .o. lets see

```python
genre_sales = (cleaned_df.groupby("Genre")["Global_Sales"].sum().sort_values(ascending=False))
```

```python
genre_sales
```

Action games generate the highest total global revenue, that makes sense XD

### Q4. Who are the dominant players in the video game industry based on total sales?

```python
pub_sales = (cleaned_df.groupby("Publisher")["Global_Sales"].sum().sort_values(ascending=False))

pub_sales.head(10)
```

As expected, its Nintendo

### Q5. Which games sold more than 10 million copies globally?

```python
ten_mil_copies = cleaned_df[cleaned_df["Global_Sales"] > 10].sort_values(by="Global_Sales", ascending=False)[["Name", "Platform", "Global_Sales"]]
```

```python
ten_mil_copies.head(10)
```

Only a small number of titles cross the 10-million-sales mark :/

### Q6: How many games were released for the PlayStation 2 (PS2) platform?

```python
ps2_count = cleaned_df[cleaned_df['Platform'] == 'PS2'].shape[0]
ps2_count
```

### Q7: Which games resonated more strongly with European audiences than North American audiences?

```python
cleaned_df[cleaned_df['EU_Sales'] > cleaned_df['NA_Sales']].sort_values('Global_Sales', ascending = False).head(5)[['Name', 'EU_Sales', 'NA_Sales', 'Global_Sales']]
```

The games are Nintendogs, GTA V, Brain Age, Brain Age 2 as well as COD:BO3

### Q8: What is the average global sale for action games vs strategy games?

```python
avg_action = cleaned_df[cleaned_df['Genre'] == 'Action']['Global_Sales'].mean()
avg_strategy = cleaned_df[cleaned_df['Genre'] == 'Strategy']['Global_Sales'].mean()
print(f"Average Global Sales   Action: {avg_action:.2f}M, Strategy: {avg_strategy:.4f}M")
```

Average Global 
Sales Action: 0.53M, Strategy: 0.2572M
